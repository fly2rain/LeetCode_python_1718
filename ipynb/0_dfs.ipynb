### [200.med-Number of Islands](https://leetcode.com/problems/number-of-islands/description/)
##### é¢˜æ„: ç»™å®šçŸ©é˜µ, åªåŒ…å« 0 æˆ– 1, 1 è¡¨ç¤º é™†åœ°, æ±‚ æ€»å…±æœ‰å¤šå°‘ä¸ªç‹¬ç«‹çš„ island?

##### æ€»ç»“: æ‰¾åˆ° island, DFS æ‰¾åˆ°ä¸º  1 çš„æ‰€æœ‰å…ƒç´ , å°†è¿™äº›å…ƒç´ è®¾ä¸ºé›¶.

```python
"""
Example 1, answer: 1 ------------------------------------
11110
11010
11000
00000

Example 2, answer: 3 ------------------------------------
11000
11000
00100
00011
"""
```

```python
# â¤ï¸ åŸæ¥å¯ä»¥åœ¨çŸ©é˜µä¸­ dfs çš„.
class Solution(object):
    def numIslands(self, grid):
        """
        :type grid: List[List[str]]
        :rtype: int
        """
        if not grid or (len(grid)==1 and not grid[0]): return 0
        
        rst = 0
        nr, nc = len(grid), len(grid[0])
        for r in xrange(nr):
            for c in xrange(nc):
                rst += grid[r][c] == "1" # â¤ï¸
                self.dfs(grid, r, c, nr, nc)
        return rst
                
    def dfs(self, grid, r, c, nr, nc): # â¤ï¸ å°†ä¸å½“å‰"1"è¿æ¥çš„æ‰€æœ‰èŠ‚ç‚¹è®¾ä¸º"0
        # 1 edge condition
        if r<0 or c<0 or r>=nr or c>=nc: return 
        if grid[r][c] == "0": return 
        
        # 2 do the dfs operation
        grid[r][c] = "0"
        self.dfs(grid, r-1, c, nr, nc)
        self.dfs(grid, r+1, c, nr, nc)
        self.dfs(grid, r, c-1, nr, nc)
        self.dfs(grid, r, c+1, nr, nc)  
```

### [124.hard-Binary Tree Maximum Path Sum](https://leetcode.com/problems/binary-tree-maximum-path-sum/description/)
##### é¢˜æ„: æ‰¾åˆ° binary tree ä»»æ„èŠ‚ç‚¹ä¹‹é—´ path çš„ æœ€å¤§ sum.

```python
class Solution(object):
    def __init__(self):
        self.maxSum = -1<<31
        
    def maxPathSum(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        if not root: return 0
        self.dfs(root)
        return self.maxSum
        
    def dfs(self, root):
        if not root: return 0
        
        left = max(self.dfs(root.left), 0) # ğŸ‘ğŸ‘ â¤ï¸â¤ï¸ 
        right = max(self.dfs(root.right), 0) # ğŸ‘ğŸ‘ â¤ï¸â¤ï¸ 
        
        self.maxSum = max(self.maxSum,  root.val + left + right)
        return root.val + max(left, right)        
```
