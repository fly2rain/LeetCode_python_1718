### [‚ù§Ô∏è96.med-Unique Binary Search Trees](https://leetcode.com/problems/unique-binary-search-trees/description/)
##### È¢òÊÑè: Given n, how many structurally unique BST's (binary search trees) that store values 1...n?
* È¢òÊÑèÁöÑÂÖ≥ÈîÆÊòØ BST, ÊâÄ‰ª•ÊûÅÂ§ßÁöÑÈôêÂà∂‰∫ÜÊ†ëÁöÑÂèØËÉΩÊÄß, ÊûÑÈÄ†ÁöÑüå≤ ÂøÖÈ°ªÊª°Ë∂≥ BST ÁöÑÊÄßË¥®.
* Dynamic programming ÈóÆÈ¢ò.

```python
"""
 * 96. Unique Binary Search Trees
 * Given n, how many structurally unique BST's (binary search trees) that store values 1...n?
     For examplse,
     Given n = 3, there are a total of 5 unique BST's.
     1         3     3      2      1
      \       /     /      / \      \
       3     2     1      1   3      2
      /     /       \                 \
     2     1         2                 3
     n = 3
     root : 1   left : 0 right : 2   f(0) * f(2);
     root : 2   left : 1 right : 1   f(1) * f(1);
     root : 3   left : 2 right : 0   f(2) * f(0);
     f(n) = f(0)*f(n-1) + f(1)*f(n-2) + ... + f(n-2)*f(1) + f(n-1)*f(0) #‚ù§Ô∏èüëçüëç 0 Âíå n -1 Ë°®Á§∫Â∑¶Âè≥Â≠êÊ†ëÁöÑËäÇÁÇπ‰∏™Êï∞
     time : O(n);
     space : O(n);
 * @param n
 * @return
"""
```

```python
# f(n) = f(0)*f(n-1) + f(1)*f(n-2) + .... f(n-1)*f(0)
class Solution(object):
    def numUniqueBST(self, n):
        f = [0] * (n + 1)
        f[0] = 1
        for i in xrange(1, n+1):
            for j in xrange(i):
                f[i] += f[j] * f[i-j+1]
        return f[n]
```

### [‚ù§Ô∏è95.med-Unique Binary Search Trees II](https://leetcode.com/problems/unique-binary-search-trees-ii/description/)
##### È¢òÊÑè: Given an integer n, generate all structurally unique BST's (binary search trees) that store values 1...n.
* 96 È¢òÊòØËÆ°Êï∞, 95 È¢òÊòØÈÅçÂéÜÊâÄÊúâÂèØËÉΩÁöÑÁªìÊûÑ. ÈöæÂ∫¶Â§ß‰∫ÜÈùûÂ∏∏Â§ö.

```python
class Solution(object):
    def buildTree(self, n):
        if n < 0: return []
        return self.helper(1, n)
        
    def helper(self, start, end):
        rst = []
        if start > end:
            rst.append(None)
            return rst
        
        for i in xrange(start, end+1):
            leftTree = self.helper(start, i-1)
            rightTree = self.helper(i+1, end)
            for left in leftTree:
                for right in rightTree:
                    node = TreeNode(i)
                    node.left = left
                    node.right = right
                    rst.append(node)
        return rst
```

### [99. Recover Binary Search Tree](https://leetcode.com/problems/recover-binary-search-tree/description/)
##### È¢òÊÑè: Two elements of a binary search tree (BST) are swapped by mistake.
* Recover the tree without changing its structure.
* È¢ùÂ§ñË¶ÅÊ±Ç: O(n) space is pretty straight forward. Could you devise a constant space solution? constant space ÈúÄË¶Å morris traversal, Â§™Èöæ‰∫Ü.

```python
class Solution(object):
    def recoverBST(self, root):
        if not root: return 
        
        prev = first = second = None
        stack, node = [], root
        while stack or node:
            while node:
                stack.append(node)
                node = node.left
            
            node = stack.pop()
            if prev and prev.val > node.val:
                if not first:
                    first = prev
                second = node     
                
            prev = node
            node = node.right
            
        first.val, second.val = second.val, first.val
```

### [450.med-Delete Node in a BST](https://leetcode.com/problems/delete-node-in-a-bst/description/)
##### È¢òÊÑè: ÁªôÂÆö BST Ê†πËäÇÁÇπÂíå key, Âà†Èô§ BST ‰∏≠ÊâÄÊúâkey Á≠â‰∫éÁªôÂÆöÂÄºÂæó node.

```python
class TreeNode(object):
    def __init__(self, val):
        self.left = None
        self.right = None
        self.val = val
    
class BST(object):
    def __init__(self):
        self.root = None
        
    # ‚ù§Ô∏è‚ù§Ô∏è insert ------------------------------- ‚ù§Ô∏è‚ù§Ô∏è
    # Ë¶ÅÁÇπ: 1) ÈÅáÂà∞ None node ÂàõÂª∫, 2) left, right Ë°îÊé•Â•Ω, 3) ËøîÂõûÁªìÊûú.
    def insertNode(self, root, val):
        if not root: 
            return TreeNode(val)
        
        if val < root.val:
            root.left = self.insertNode(root.left, val)
        else:
            root.right = self.insertNode(root.right, val)
        return root
        
    def deleteNode(self, root, key):
        if not root: return None
        
        if key < root.val:
            root.left = self.deleteNode(root.left, key)
        elif key > root.val:
            root.right = self.deleteNode(root.right, key)
        else:
            if not root.left:
                return root.right
            elif not root.right:
                return root.left
            else:
                succ = root.right
                while succ.left:
                    succ = succ.left
                root.val = succ.val
                root.right = self.deleteNode(root.right, root.val)
        return root        
```

### [114.med-Flatten Binary Tree to Linked List](https://leetcode.com/problems/flatten-binary-tree-to-linked-list/description/)
##### È¢òÊÑè: ÁªôÂÆö binary tree, in-place ËΩ¨ÂåñÊàê linked list.

```python
"""
Given 
         1
        / \
       2   5
      / \   \
     3   4   6
The flattened tree should look like:
       1
        \
         2
          \
           3
            \
             4
              \
               5
                \
"""
```

```python
class Solution(object):
    def __init__(self):
        self.prev = None
        
    def flattenTree(self, root):
        if not root: return 
        
        self.flattenTree(root.right)
        self.flattenTree(root.left)
        
        root.right = self.prev
        root.left = None
        self.prev = root
```

### [98.med-Validate Binary Search Tree](https://leetcode.com/problems/validate-binary-search-tree/description/)
##### È¢òÊÑè: ÁªôÂÆö binary tree, Âà§Êñ≠ÂÆÉÊòØÂê¶ÊòØ BST?

##### ÂàÜÊûê
* ÁªôÂÆö binary tree, Êó†ÈùûÂ∞±ÊòØÂà§Êñ≠ left <= root <= right? Ëøò‰∏çÂ§üÁöÑ, <span class="girk">‰∏ç‰ªÖÈúÄË¶ÅÂØπÊØî left/right node Âíå current node ÂÖ≥Á≥ª, ËøòË¶ÅÂØπÊØî left sub-tree Âíå current node ÂÖ≥Á≥ª.</span> Âõ†‰∏∫ÊúâÊó∂, ËôΩÁÑ∂ left node Á¨¶ÂêàÊù°‰ª∂, ‰ΩÜÂÆÉÁöÑÂ≠êÊ†ë‰∏≠ËäÇÁÇπÊúâ‰∏çÊª°Ë∂≥Êù°‰ª∂. 
* Ê†πÊçÆÊù°‰ª∂, ÊòØ left-subtree <= root <= right-subtree.

##### ÂèÇËÄÉËßÜÈ¢ë:
* Ëä±Ëä±: https://www.youtube.com/watch?v=Jq0Wk9xeQ0U ËÆ≤ÂæóÈùûÂ∏∏Ê∏ÖÊ•ö. ÈÄöËøá

```python
class Solution(object):
    def isBST(self, root):
        return self.helper(root, float("-inf"), float("inf"))
        
    def helper(self, root, left, right):
        if not root: return True
        
        return (
            left < root.val < right and 
            self.helper(root.left, left, root.val) and 
            self.helper(root.right, root.val, right)
        )
```

### [536.med-Construct Binary Tree from String](https://leetcode.com/problems/construct-binary-tree-from-string/description/)
##### È¢òÊÑè: Âíå  606.easy-Construct String from Binary Tree Áõ∏Âèç. ÁªôÂÆö string ÊûÑÈÄ†Âá∫ binary tree.
* <span class="girk">String ÁöÑÁâπÁÇπ</span>: 
    * has zero, one or two pairs of parenthesis.
    * The integer represents the root's value and a pair of parenthesis contains a child binary tree with the same structure.
    * You always start to construct the left child node of the parent first if it exists.
* Note that:
    * There will only be '(', ')', '-' and '0' ~ '9' in the input string.
    * An empty tree is represented by "" instead of "()".

```python
""" Example 1 ***************************************************************
Input: Binary tree: [1,2,3,4]
       1
     /   \
    2     3
   /    
  4     
  
Output: "1(2(4))(3)"

Explanation: Originallay it needs to be "1(2(4)())(3()())", 
but you need to omit all the unnecessary empty parenthesis pairs. 
And it will be "1(2(4))(3)"
"""

""" Example 2 ***************************************************************
Input: Binary tree: [1,2,3,null,4]
       1
     /   \
    2     3
     \  
      4 

Output: "1(2()(4))(3)"

Explanation: Almost the same as the first example, 
except we can't omit the first parenthesis pair to break the one-to-one mapping 
relationship between the input and the output.
"""
```

```python
class Solution(object):
    ## 606 -----------------------
    def constructString(self, root):
        if not root: return ""
        
        cur = str(root.val)
        left = self.constructString(root.left)
        right = self.constructString(root.right)
        if not left and not right:
            return cur
        if not right:
            return cur + "(" + left + ")"
        
        return cur + "(" + left + ")(" + right + ")"
    
    ## 536 -----------------------
    def buildTree(self, s):
        if not s: return None
        
        i = s.find("(")
        if i < 0: return TreeNode(int(s))
        
        cnt = 0
        for j, x in enumerate(s):
            if x == "(":
                cnt += 1
            elif x == ")":
                cnt -= 1
            
            if cnt == 0 and j > i:
                break
                
        root = TreeNode(int(s[:i]))
        root.left = self.buildTree(s[i+1:j])
        root.right = self.buildTree(s[j+2:-1])
        return root
```

### [297.hard-Serialize and Deserialize Binary Tree](https://leetcode.com/problems/serialize-and-deserialize-binary-tree/description/)
##### È¢òÊÑè: Â∞Ü binary tree ËΩ¨ÂåñÊàê string (‰æø‰∫é storage Âíå ‰º†Ëæì), ÂèçËøáÊù•Â∞Ü string ËΩ¨ÂåñÊàê binary tree.

##### ÊÄùË∑Ø: BFS serialize Âíå deserialize. 

##### ÂèÇËÄÉËßÜÈ¢ë:
* Edward Shi: https://www.youtube.com/watch?v=8UKHhUhXugo
* huahua: https://www.youtube.com/watch?v=8UKHhUhXugo

```python
from collections import deque
class codeBinaryTree(object):
    def serialize(self, root):
        """Encodes a tree to a single string.
        :type root: TreeNode
        :rtype: str
        """
        if not root: return ""
        
        rst = []
        q = deque([root])
        while q:
            node = q.popleft()
            if node is None:
                rst.append("#")
                continue
                
            rst.append(str(root.val))           
            q.append(node.left)
            q.append(node.right)
            
        return " ".join(q)
        
    def deSerialize(self, data):
        if not data: return None
        
        s = deque(data.split(" "))
        root = TreeNode(int(s.popleft()))
        q = deque([root])
        while s:
            node = q.popleft()
            
            x = s.popleft()
            if x != "#":
                node.left = TreeNode(int(x))
                q.append(node.left)
                
            x = s.popleft()
            if x != "#":
                node.right = TreeNode(int(x))
                q.ppend(node.right)
        return root
```

### [449.med-Serialize and Deserialize BST](https://leetcode.com/problems/serialize-and-deserialize-bst/description/)
##### È¢òÊÑè: ÂØπ BST ÁöÑ serialize & deserialize.

##### <span class="burk">ÊÄªÁªì</span>: Âà©Áî® BST preorder ÁöÑÁâπÁÇπ, ËäÇÁúÅÂ§ßÈáèÁöÑÁ©∫Èó¥. <span class="girk">root -> left -> right</span>, ÂÖ∂‰∏≠ <span class="girk">left ÊÄªÊòØÂ∞è‰∫é root</span>, Âà©Áî®Ëøô‰∏™ÊÄßË¥®Ëá™ÁÑ∂<span class="girk">Âå∫ÂàÜÂºÄ left tree Âíå right tree</span>.

```python
from collections import deque
class codeBST(object):
    def serialize(self, root):
        if not root: return ""
        
        rst = []
        stack = [root]
        while stack:
            node = stack.pop()
            
            rst.append(str(node.val))
            if node.right: stack.append(node.right)
            if node.left: stack.append(node.left)
                
        return " ".join(rst)
    
    def deSerialize(self, data):
        if not data: return None
        
        q = deque()
        for x in data.split(" "): 
            q.append(int(x))
            
        return self.helper(q)
            
    def helper(self, q):
        if not q: return None
        
        node = TreeNode(q.popleft())
        
        small_q = deque()
        while q and node.val > q[0]:
            small_q.append(q.popleft())
        
        node.left = self.helper(small_q)
        node.right = self.helper(q)
        return node
```

### [‚ù§Ô∏è652.med-Find Duplicate Subtrees](https://leetcode.com/problems/find-duplicate-subtrees/description/)
##### È¢òÊÑè: Given a binary tree, return all duplicate subtrees. For each kind of duplicate subtrees, you only need to return the root node of any one of them.
* Two trees are duplicate if they have the same structure with same node values.

```python
"""
Example 1: 
        1
       / \
      2   3
     /   / \
    4   2   4
       /
      4
      
The following are two duplicate subtrees:
      2
     /
    4
and
    4
    
Therefore, you need to return above trees' root in the form of a list.
"""
```

```python
class TrieNode(object):
    def __init__(self):
        self.word = ""
        self.dic = {}
        
class Solution(object):
    def findWords(self, board, words):
        """
        :type board: List[List[str]]
        :type words: List[str]
        :rtype: List[str]
        """
        if not board or (len(board)==1 and not board[0]) or not words: return []
        
        root = self.buildTrie(words)
        rst = []
        nr, nc = len(board), len(board[0])
        for r in xrange(nr):
            for c in xrange(nc):
                self.helper(board, nr, nc, r, c, root, rst)
        return rst                
    
    def helper(self, board, nr, nc, r, c, root, rst):
        if r<0 or c<0 or r>=nr or c>=nc: return 
        
        x = board[r][c]
        if x=="#": return 
        if x not in root.dic: return
        
        root = root.dic[x]
        if root.word:
            rst.append(root.word)
            root.word = ""
        
        board[r][c] = "#"
        self.helper(board, nr, nc, r-1, c, root, rst)
        self.helper(board, nr, nc, r+1, c, root, rst)
        self.helper(board, nr, nc, r, c-1, root, rst)
        self.helper(board, nr, nc, r, c+1, root, rst)
        board[r][c] = x
        
    def buildTrie(self, words):
        root = TrieNode()
        for word in words:
            node = root
            for x in word:
                if x not in root.dic:
                    root.dic[x] = TrieNode()
                root = root.dic[x]
            root.word = word
        return root
```

```python
class Solution(object):
    def countBattleships(self, board):
        """
        :type board: List[List[str]]
        :rtype: int
        """
        if not board or (len(board)==1 and not board[0]): return 0
        
        rst = 0
        nr, nc = len(board), len(board[0])
        for r in xrange(nr):
            for c in xrange(nc):
                if board[r][c] == ".": 
                    continue
                if r>0 and board[r-1][c] == "X":
                    continue
                if c>0 and board[r][c-1] == "X":
                    continue
                rst += 1
        return rst
    
    # ‚ù§Ô∏è‚ù§Ô∏è‚ù§Ô∏è‚ù§Ô∏è‚ù§Ô∏è
    def uniquePaths(self, m, n): # ‚ù§Ô∏è time O(m*n), space O(min(m, n))
        """
        :type m: int
        :type n: int
        :rtype: int
        """
        if m < n:
            return self.uniquePaths(n, m)
        
        dq = [0] * (n)
        dp[0] = 1
        for i in xrange(m):
            for j in xrange(1, n):
                 # dp[i][j] = dp[i-1][j] + dp[i][j-1]
                dp[j] = dp[j] + dp[j-1] # ‚ù§Ô∏è‚ù§Ô∏èdp[j] Ë°®Á§∫‰∏ä‰∏ÄË°åÁöÑ j, dp[j-1] Ë°®Á§∫Ëøô‰∏ÄË°åÁöÑ j-1 Âàó
        return dp[n-1]
      
    # ‚ù§Ô∏è‚ù§Ô∏è‚ù§Ô∏è‚ù§Ô∏è‚ù§Ô∏è
    def uniquePathsWithObstacles(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: int
        """
        if not grid or (len(grid)==1 and not grid[0]): return 0
        
        nr, nc = len(grid), len(grid[0])
        dp = [0]*nc
        dp[0] = 1 # ‚ù§Ô∏è‚ù§Ô∏è Ëøô‰∏™Êù°‰ª∂ÂÆπÊòìÊºè‰∫Ü
        for r in xrange(nr):
            for c in xrange(nc):
                if grid[r][c] == 1: dp[c] = 0
                if c>0:
                    dp[c] = dp[c] + dp[c-1]
        return dp[nc-1]
    
    # ‚ù§Ô∏è‚ù§Ô∏è‚ù§Ô∏è‚ù§Ô∏è‚ù§Ô∏è#‚ù§Ô∏è‚ù§Ô∏è‚ù§Ô∏è‚ù§Ô∏è‚ù§Ô∏è
    def minPathSum(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: int
        """
        if not grid or (len(grid)==1 and not grid[0]): 
            return 0
        
        nr, nc = len(grid), len(grid[0])
        for r in xrange(nr):
            for c in xrange(nc):
                if r==0 and c!=0:
                    grid[r][c] += grid[r][c-1]
                elif c==0 and r!=0:
                    grid[r][c] += grid[r-1][c]
                elif r!=0 and c!=0:
                    grid[r][c] += min(grid[r-1][c], grid[r][c-1])
        return grid[nr-1][nc-1]
    
    
    def isValidSudoku(self, board):
        """
        :type board: List[List[str]]
        :rtype: bool
        """
        n = len(board)
        seen = set()
        for r in xrange(n):
            for c in xrange(n):
                x = board[r][c]
                if x == ".":
                    continue
                if (r,x) in seen or (x,c) in seen or (r/3,c/3,x) in seen:
                    return False
                seen.add((r,x))
                seen.add((x,c))
                seen.add((r/3,c/3,x))
        return True
```
