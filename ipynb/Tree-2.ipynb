```python
# TC & SC: O(n), 77ms. åŠ é€Ÿ 5 å€ ä»…ä»…å› ä¸ºä½¿ç”¨ dict å­˜å‚¨ inorder, é¿å…äº†æ¯æ¬¡åœ¨ inorder ä¸­æœç´¢.
class Solution(object):
    def buildTree(self, preorder, inorder):
        """
        :type preorder: List[int]
        :type inorder: List[int]
        :rtype: TreeNode
        """
        if not preorder or not inorder or len(preorder) != len(inorder): 
            return None
        
        indic = {}
        for i, x in enumerate(inorder): 
            indic[x] = i
        return self.helper(0, len(inorder)-1, 0, len(inorder), indic, preorder)
        
        
    def helper(self, preLow, preHigh, inLow, inHigh, indic, preorder):
        if preLow > preHigh or inLow > inHigh: return None
        
        node = TreeNode(preorder[preLow])
        inIdx = indic[node.val]
        
        node.left = self.helper(preLow+1, preLow+(inIdx-inLow), inLow, inIdx-1, 
                               indic, preorder)
        node.right = self.helper(preLow+(inIdx-inLow)+1, preHigh, inIdx+1, inHigh,
                                indic, preorder)
        return node
```

### [â¤ï¸106.medium-Construct Binary Tree from Inorder and Postorder Traversal](https://leetcode.com/problems/construct-binary-tree-from-inorder-and-postorder-traversal/description/)
##### é¢˜æ„: ç»™å®š postorder å’Œ inorder çš„ list representation, æ„é€  binary tree.
* postorder: left -> right -> root
* inorder: left -> root -> right

##### æ€»ç»“
* å’Œ 105.medium-Construct Binary Tree from Preorder and Inorder Traversal å¤§åŒå°å¼‚.
 

```python
class Solution(object):
    def buildTree(self, inorder, postorder):
        """
        :type inorder: List[int]
        :type postorder: List[int]
        :rtype: TreeNode
        """
        if not inorder or not postorder or len(inorder) != len(postorder):
            return None
        
        indic = {}
        for i, x in enumerate(inorder):
            indic[x] = i
        return self.helper(indic, postorder, 0, len(inorder)-1, 0, len(inorder)-1)
    
    def helper(self, indic, postorder, inLow, inHigh, poLow, poHigh):
        if poLow > poHigh or inLow > inHigh:
            return None
        
        node = TreeNode(postorder[poHigh])
        inIdx = indic[node.val]

        left_length = inIdx - inLow
        node.left = self.helper(indic, postorder, inLow, inIdx-1,
                                poLow, poLow+left_length-1)
        node.right = self.helper(indic, postorder, inIdx+1, inHigh, #â å†™æˆ poHigh, 
                                 poLow+left_length, poHigh-1) # è€Œä¸æ˜¯ poHigh - 1
        return node
```

```python
# æ–¹æ³•1: BFS æ¯ä¸€å±‚ right -> left æ¸¸èµ°, åˆ°äº†æœ€åä¸€å±‚, è‡ªç„¶ the leftmost å…ƒç´ æ˜¯æœ€åçš„.
from collections import deque
class Solution(object):
    def findBottomLeftValue(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        # 1 edge condtion
        if not root: return None
        
        # 2 find it via BFS iteration
        queue, rst = deque([root]), None        
        while queue:
            node = queue.popleft()
            rst = node.val
            
            if node.right: queue.append(node.right) # ğŸ‘ğŸ‘ğŸ‘ ç‚¹ç›ä¹‹ç¬”åœ¨äºå…ˆå³åå·¦, ä¿è¯æœ€åä¸€ä¸ªå…ƒç´ 
            if node.left: queue.append(node.left) # æ˜¯æœ€åä¸€å±‚çš„æœ€å·¦è¾¹é‚£ä¸ªå…ƒç´ .
        return rst
```

```python
# æ–¹æ³•2: BFS æ¯ä¸€å±‚ left -> right æ¸¸èµ°, æœ€åä¸€å±‚, è®°å½•ç¬¬ä¸€ä¸ª(å³æ˜¯ left most), ä½†éœ€è¦ä¸€ä¸ªè®°å½•å˜é‡ i == 0.
from collections import deque
class Solution(object):
    def findBottomLeftValue(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        # 1 edge condtion
        if not root: return None
        
        # 2 find it via BFS iteration
        queue, rst = deque([root]), None        
        while queue:
            n = len(queue)
            for i in xrange(n):
                node = queue.popleft()
                if i == 0: rst = node.val
                    
                if node.left: queue.append(node.left) # âââ ç«Ÿç„¶æŠŠè¿™ä¸¤æ¡è¯­å¥å†™åˆ° for å¾ªç¯å¤–é¢äº†.
                if node.right: queue.append(node.right)  
                
        return rst
```

```python
# æ–¹æ³•3: DFS é€šè¿‡è¾…åŠ©å˜é‡ height æ ‡å®šæœ€æ–°çš„ depth, å¦‚æœæœ‰æ›´æ·±çš„ height, åˆ™è¡¨æ˜è¿™æ˜¯æœ€åä¸€å±‚. æ¯”è¾ƒå·§å¦™çš„æ˜¯,
# è¿™é‡Œåªä¿ç•™ç¬¬ä¸€æ¬¡åˆ°è¾¾æœ€é«˜ height çš„å€¼. æ‰€ä»¥, preorder, inorder, postorder éƒ½å¯ä»¥, å› ä¸ºéƒ½èƒ½ä¿è¯ left ä¼˜å…ˆäº right

class Solution(object):
    def __init__(self):
        self.height, self.rst = 0, None
        
    def findBottomLeftValue(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        # 1 edge case 
        if not root: return None
        
        # 2 do the coputation
        self.dfs(root, 1) # â è¾“å…¥çš„æ˜¯ 1, å¯ root æ˜¯ç¬¬ä¸€å±‚.
        return self.rst
        
    def dfs(self, node, depth): # ğŸ‘ğŸ‘ğŸ‘ åªè¦æ˜¯ dfs, å‰åº ä¸­åº å’Œ ååºå€’æ— æ‰€è°“, åªè¦ left å…ˆäº right
        if not node: return 
        
        if depth > self.height: # ğŸ‘ğŸ‘ğŸ‘ ç‚¹ç›ä¹‹ç¬”, åªèƒ½è€ƒè™‘ç¬¬ä¸€æ¬¡å‡ºç°çš„ node.
            self.height = depth
            self.rst = node.val
            
        self.dfs(node.left, depth + 1)
        self.dfs(node.right, depth + 1)
```

```python
# æ–¹æ³•1: æ˜¾ç„¶å¯ä»¥ä½¿ç”¨ BFS æ¸¸èµ°æ¯ä¸€è¡Œ, å¯¹æ¯”å‡ºæœ€å¤§çš„å€¼..
from collections import deque
class Solution(object):
    def largestValues(self, root):
        """
        :type root: TreeNode
        :rtype: List[int]
        """
        # 1 edge condition
        if not root: return []
        
        # 2 find them
        rst = []
        queue = deque([root]) #â¤ï¸â¤ï¸ BFS ä½¿ç”¨ queue å°±å¯ä»¥äº†, ä¸éœ€è¦ç©æŠ€å·§ä½¿ç”¨ stack.
        while queue:
            n, maxval = len(queue), float("-inf")
            for i in xrange(n):
                node = queue.popleft()
                maxval = max(maxval, node.val)
                if node.left: queue.append(node.left) # âğŸ‘â¤ï¸ å¾ˆå®¹æ˜“æŠŠ append è¿™ä¸¤å¥å†™åˆ° for å¾ªç¯å¤–é¢.
                if node.right: queue.append(node.right)
            rst.append(maxval)
            
        return rst
```

```python
# æ–¹æ³•2: DFS å¯ä»¥å—? ä¹Ÿåº”è¯¥å¯ä»¥, éœ€è¦ä¸€ä¸ªè¾…åŠ©å˜é‡è®°å½•å½“å‰ depth, å¯¹æ¯” depth ä¸€æ ·çš„ element çš„å…ƒç´ å¤§å°, è¿”å›æœ€å¤§çš„é‚£ä¸ª
class Solution(object):
    def __init__(self):
        self.rst = []
        
    def largestValues(self, root):
        """
        :type root: TreeNode
        :rtype: List[int]
        """
        self.dfs(root, 0)
        return self.rst
        
    def dfs(self, node, depth):
        if not node: return 
        
        while len(self.rst) < depth + 1: # â æ³¨æ„ depth + 1, depth æ˜¯ index ä¸æ˜¯è®¡æ•°.
            self.rst.append(float("-inf"))
        self.rst[depth] = max(self.rst[depth], node.val)
        
        self.dfs(node.left, depth + 1)
        self.dfs(node.right, depth + 1)
```

* æ³¨æ„: è¿™é‡Œçš„ width æ¯”è¾ƒç‰¹åˆ«. The width of one level is defined as the length <span class="girk">between the end-nodes</span> (the leftmost and right most non-null nodes in the level, where the null nodes between the end-nodes are also counted into the length calculation.

    * ç¬¬ä¸€å±‚ index æ˜¯ $2*0$
    * ç¬¬äºŒå±‚ index æ˜¯ $2*0$ for left, $2*0 + 1=1$ for right
    * ç¬¬ä¸‰å±‚ index æ˜¯ $2*0$ for left.left, $2*0 + 1$ for left.right, $2*1$ for right.left, $2*1+1$ for right.right
    
* <span class="girk">æ–¹æ³•2: DFS ç¨å¾®éº»çƒ¦ä¸€äº›, ä¹Ÿèƒ½åš. éœ€è¦å¾ˆå¤šè¾…åŠ©å˜é‡.</span>

##### æ€»ç»“: ä¹Ÿå¯ä»¥ä½¿ç”¨è¿™ç§æ ‡å‡†çš„ index æ–¹æ³•.
* é¦–å…ˆæ˜ç¡® binary tree å’Œ Array representation of Binary tree çš„å…³ç³».
* ä¸€ç§æ¯”è¾ƒé€šç”¨æ— æ­§ä¹‰çš„æ–¹å¼æ˜¯ BFS. å‚è€ƒ: https://en.wikipedia.org/wiki/Binary_tree
    * Binary trees can also be stored in breadth-first order as an implicit data structure in arrays, and if the tree is a complete binary tree, this method wastes no space. 
    * <span class="burk">parent index --> children index</span>: if a node has an index i, its children are found at indices <span class="girk">${\displaystyle 2i+1}$ (for the left child)</span> and <span class="girk">${\displaystyle 2i+2}$ (for the right)</span>,
    * <span class="burk">children index --> parent index</span>: while its parent (if any) is found at index ${\displaystyle \left\lfloor {\frac {i-1}{2}}\right\rfloor }$  (assuming the root has index zero).

```python
# æ–¹æ³•1: ä½¿ç”¨ BFS list æ ‡å‡†çš„ representation, ä½¿ç”¨ queue, å­˜æ”¾æ¯ä¸ª node
# çš„æ—¶å€™æŠŠ node çš„ index ä¹Ÿå­˜è¿›å».
from collections import deque
class Solution(object):
    def widthOfBinaryTree(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        # edge condition
        if not root: return 0
        
        maxWidth = 0
        queue = deque([(root, 0)])  # ğŸ‘ğŸ‘ğŸ‘,æ·»åŠ  node + index
        while queue:
            n = len(queue)
            # ââ è¿™æ”¾åœ¨ for å¾ªç¯ä¹‹åå°±å‡ºé”™äº†. ğŸ‘ğŸ‘ æ¯æ¬¡æ„é€ ç© queue ä¹‹å, æµ‹é‡é•¿åº¦. å°±åƒ n = len(queue) ä¸€æ ·.
            maxWidth = max(maxWidth, queue[-1][1] - queue[0][1] + 1) # ğŸ‘ğŸ‘ğŸ‘ deque ä¹Ÿæœ‰ [-1] çš„æ–¹ä¾¿ç´¢å¼•.
            for i in xrange(n):
                node, ind = queue.popleft()
                if node.left: queue.append((node.left, 2*ind + 1)) # ğŸ‘ğŸ‘ğŸ‘,æ·»åŠ  node + index
                if node.right: queue.append((node.right, 2*ind + 2))
        return maxWidth             
```

```python
# æ–¹æ³•2: BFS, ä½¿ç”¨ queue, å­˜æ”¾æ¯ä¸ª node çš„æ—¶å€™æŠŠ node çš„ index ä¹Ÿå­˜è¿›å».
from collections import deque
class Solution(object):
    def widthOfBinaryTree(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        # edge condition
        if not root: return 0
        
        maxWidth = 0
        queue = deque([(root, 0)])  # ğŸ‘ğŸ‘ğŸ‘,æ·»åŠ  node + index
        while queue:
            n = len(queue)
            # ââ è¿™æ”¾åœ¨ for å¾ªç¯ä¹‹åå°±å‡ºé”™äº†. ğŸ‘ğŸ‘ æ¯æ¬¡æ„é€ ç© queue ä¹‹å, æµ‹é‡é•¿åº¦. å°±åƒ n = len(queue) ä¸€æ ·.
            maxWidth = max(maxWidth, queue[-1][1] - queue[0][1] + 1) # ğŸ‘ğŸ‘ğŸ‘ deque ä¹Ÿæœ‰ [-1] çš„æ–¹ä¾¿ç´¢å¼•.
            for i in xrange(n):
                node, ind = queue.popleft()
                if node.left: queue.append((node.left, 2*ind)) # ğŸ‘ğŸ‘ğŸ‘,æ·»åŠ  node + index
                if node.right: queue.append((node.right, 2*ind + 1))
        return maxWidth             
```

```python
# æ–¹æ³•3: DFS ç¨å¾®éº»çƒ¦ä¸€äº›, ä¹Ÿèƒ½åš. éœ€è¦å¾ˆå¤šè¾…åŠ©å˜é‡
Solution(object):
    def __init__(self):
        self.nums, self.maxwidth = [], 0 # self.num å­˜æ”¾æ¯ä¸ª level çš„ min index å’Œ max index
        
    def widthOfBinaryTree(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        if not root: return 0
        self.dfs(root, 0, 0)
        return self.maxwidth
        
    def dfs(self, node, depth, ind):
        # edge case
        if not node: return 
        
        if 1 + depth > len(self.nums): # ââ åŸæ¥å†™çš„æ˜¯ 1 + depth, æ³¨æ„ depth æ˜¯ index ä¸æ˜¯è®¡æ•°, æ‰€ä»¥è¦åŠ  1.
            self.nums.append([float("inf"), 0])
            
        self.nums[depth][0] = min(self.nums[depth][0], ind)
        self.nums[depth][1] = max(self.nums[depth][1], ind)
        if self.nums[depth][0] != float("inf"):
            self.maxwidth = max(self.maxwidth, self.nums[depth][1] - self.nums[depth][0] + 1)
        
        self.dfs(node.left, depth + 1, 2*ind)
        self.dfs(node.right, depth + 1, 2*ind + 1)
```

### [â¤ï¸96.medium-Unique Binary Search Trees](https://leetcode.com/problems/unique-binary-search-trees/description/)
##### é¢˜æ„: Given n, how many structurally unique BST's (binary search trees) that store values 1...n?
* é¢˜æ„çš„å…³é”®æ˜¯ BST, æ‰€ä»¥æå¤§çš„é™åˆ¶äº†æ ‘çš„å¯èƒ½æ€§, æ„é€ çš„ğŸŒ² å¿…é¡»æ»¡è¶³ BST çš„æ€§è´¨.
* Dynamic programming é—®é¢˜.

##### å‚è€ƒè§†é¢‘
* Edward Shi:
    * è§†é¢‘ https://www.youtube.com/watch?v=ZAq5BoTes8o
    * code: https://github.com/EdwardShi92/Leetcode-Solution-Code/blob/master/UniqueBinarySearchTrees.java

```python
"""
 * 96. Unique Binary Search Trees
 * Given n, how many structurally unique BST's (binary search trees) that store values 1...n?
     For example,
     Given n = 3, there are a total of 5 unique BST's.
     1         3     3      2      1
      \       /     /      / \      \
       3     2     1      1   3      2
      /     /       \                 \
     2     1         2                 3
     n = 3
     root : 1   left : 0 right : 2   f(0) * f(2);
     root : 2   left : 1 right : 1   f(1) * f(1);
     root : 3   left : 2 right : 0   f(2) * f(0);
     f(n) = f(0)*f(n-1) + f(1)*f(n-2) + ... + f(n-2)*f(1) + f(n-1)*f(0) #â¤ï¸ğŸ‘ğŸ‘ 0 å’Œ n -1 è¡¨ç¤ºå·¦å³å­æ ‘çš„èŠ‚ç‚¹ä¸ªæ•°
     time : O(n);
     space : O(n);
 * @param n
 * @return
"""

class Solution(object):
    def numTrees(self, n):
        """
        :type n: int
        :rtype: int
        """
        F = [0] * (n+1) #â¤ï¸ğŸ‘ ç¬¬ 0 - n ä¸ªå…ƒç´ . 
        F[0] = 1
        for i in xrange(1, n+1):
            for j in xrange(i):
                F[i] += F[j] * F[i-j-1] # â F[i] ä¸æ˜¯ F(n)
                
        return F[n]
```

### [â¤ï¸98.medium-Validate Binary Search Tree](https://leetcode.com/problems/validate-binary-search-tree/description/)
##### é¢˜æ„: ç»™å®š binary tree, åˆ¤æ–­å®ƒæ˜¯å¦æ˜¯ BST?

##### åˆ†æ
* ç»™å®š binary tree, æ— éå°±æ˜¯åˆ¤æ–­ left <= root <= right? è¿˜ä¸å¤Ÿçš„, <span class="girk">ä¸ä»…éœ€è¦å¯¹æ¯” left/right node å’Œ current node å…³ç³», è¿˜è¦å¯¹æ¯” left sub-tree å’Œ current node å…³ç³».</span> å› ä¸ºæœ‰æ—¶, è™½ç„¶ left node ç¬¦åˆæ¡ä»¶, ä½†å®ƒçš„å­æ ‘ä¸­èŠ‚ç‚¹æœ‰ä¸æ»¡è¶³æ¡ä»¶. 
* æ ¹æ®æ¡ä»¶, æ˜¯ left-subtree <= root <= right-subtree.

##### å‚è€ƒè§†é¢‘:
* èŠ±èŠ±: https://www.youtube.com/watch?v=Jq0Wk9xeQ0U è®²å¾—éå¸¸æ¸…æ¥š. é€šè¿‡

```python
# æ–¹æ³•1: up down æ–¹æ³•, è‡ªç„¶, æ˜äº†, çœå»äº†ä¼ å¤šä¸ªå‚æ•°çš„éº»çƒ¦. TC: O(n), SC: O(h)
#   1) ç»™ root è®¾ç½®ä¸€ä¸ªèŒƒå›´, å……å½“ min & max, 
#   2) preorder ä»ä¸Šåˆ°ä¸‹éå† tree èŠ‚ç‚¹, æ”¹å˜æ¯ä¸ªleft & rightèŠ‚ç‚¹çš„ min & max é€æ¸ç¼©å°èŒƒå›´.
#   3) é€’å½’è°ƒç”¨å·¦å³èŠ‚ç‚¹, æ£€æŸ¥æ¯ä¸ªèŠ‚ç‚¹æ˜¯å¦æ»¡è¶³ min < root < max.
class Solution(object):
    def isValidBST(self, root):
        """
        :type root: TreeNode
        :rtype: bool
        """
        if not root: return True
        return self.dfs(root, float("-inf"), float("inf"))
      
    def dfs(self, node, minval, maxval):
        if not node: return True
        
        return (
            minval < node.val < maxval and 
            self.dfs(node.left, minval, node.val) and 
            self.dfs(node.right, node.val, maxval)
        )
```

```python
# æ–¹æ³•2: [inorder éå†, prev < cur]
#     1. æŠ“ä½ BST inorder (left -> root -> right) æ˜¯å‡åºçš„ç‰¹ç‚¹.
#     2. æ¯æ¬¡æ£€æŸ¥ prev < cur æ˜¯å¦æ»¡è¶³.
class Solution(object): ### iteration ########################
    def isValidBST(self, root):
        """
        :type root: TreeNode
        :rtype: bool
        """
        if not root: return True
        
        stack, node, prev = [], root, float("-inf")
        while stack or node:
            while node:
                stack.append(node)
                node = node.left
                
            node = stack.pop()
            if prev >= node.val:
                return False
            
            prev = node.val
            node = node.right
            
        return True
    
class Solution(object): ### recursion ########################
    def __init__(self): #ğŸ‘ğŸ‘ğŸ‘ å› ä¸º inorder éœ€è¦è¿”å› node.val å’Œ true (false), æ‰€ä»¥å¼•å…¥æˆå‘˜å˜é‡.
        self.isBST = True
        self.prev = float("-inf")
        
    def isValidBST(self, root):
        """
        :type root: TreeNode
        :rtype: bool
        """
        if not root: return True
        self.inorder(root)
        return self.isBST
    
    def inorder(self, node):
        if not node: return float("-inf")
        
        self.inorder(node.left)
        self.isBST = self.isBST and node.val > self.prev
        self.prev = node.val
        self.inorder(node.right)
```

```python
# æ–¹æ³•3: 
#    1) bottom up é€ä¸€æ£€æŸ¥æ¯ä¸ªnode æ˜¯å¦å¤§äº left subtree ä¸­æœ€å¤§å€¼, å¹¶ä¸”å°äº right subtree æœ€å°å€¼.
#    2) æ¯ä¸ªé€’å½’è°ƒç”¨éœ€è¦è¿”å›, å·¦å­æ ‘çš„æœ€å¤§æœ€å°å€¼, å’Œ, å³å­æ ‘çš„æœ€å°æœ€å¤§å€¼. 
#    3) å¹¶ä¸”, æˆ‘ä»¬æŠŠæ¯ä¸ª node çš„ true or false åˆå¹¶åˆ°æœ€ç»ˆç»“æœä¸­. æ‰€ä»¥éœ€è¦æ¯ä¸ªé€’å½’å‡½æ•°è¿”å›ä¸‰ä¸ªæ•°å€¼. æŒºéº»çƒ¦. 
class Solution(object):
    def __init__(self):
        self.isBST = True
        
    def isValidBST(self, root):
        """
        :type root: TreeNode
        :rtype: bool
        """
        self.dfs(root)
        return self.isBST
        
    def dfs(self, node):
        if not node: 
            return float("inf"), float("-inf") # min & max
        
        l_min, l_max = self.dfs(node.left)
        r_min, r_max = self.dfs(node.right)
        
        self.isBST = self.isBST and l_max < node.val and r_min > node.val
        
        c_min = min(min(l_min, node.val), r_min)
        c_max = max(max(r_max, node.val), l_max)
        return c_min, c_max
```

```python
# è·Ÿ tree æ²¡å•¥å…³ç³», å‡€è€ƒå¯Ÿ dictionary, list sorting äº†.
# æ€è·¯:
#     1 å¯¹ letter ç»Ÿè®¡æ•°é‡, ä½¿ç”¨ hash table
#     2 å°† hash table è½¬åŒ–æˆ list of tuples, å¹¶æ ¹æ®è¯é¢‘æ’åº (ä¸ªæ€§åŒ–æ’åºè§„åˆ™ + reverse=True é™åº)
#     3 æ ¹æ®æ’åºç»“æœæ‹¼æ¥å‡ºæœ€ç»ˆç»“æœ. 
class Solution(object):
    def frequencySort(self, s):
        """ 
        :type s: str
        :rtype: str
        """
        # 1 edge condition
        if not s: return "" # â æ³¨æ„è¿”å›æ¡ä»¶.
        
        #2 ç»Ÿè®¡è¯é¢‘
        dic = {}
        for c in s:  dic[c] = dic.get(c, 0) + 1 
        
        #3 å°† dic ç»“æœè½¬åŒ–æˆ list, å¹¶ä¸ªæ€§åŒ– sort è®¾ç½®.
        freqlist = dic.items()
        freqlist.sort(key=lambda x: x[1], reverse=True) #ğŸ‘ğŸ‘ æŒ‰ç…§å…ƒç´ ç¬¬2ä¸ªå…ƒç´ , é™åº. ğŸ‘ reverse=True
        
        # 4 æ ¹æ®æ’åºç»“æœæ‹¼æ¥å‡ºæœ€ç»ˆçš„ç»“æœ. 
        rst = ""
        for x in freqlist:
            rst += x[0]*x[1] # ğŸ‘
        return rst          
```

### [508.medium-Most Frequent Subtree Sum](https://leetcode.com/problems/most-frequent-subtree-sum/description/)
##### é¢˜æ„: æ±‚è§£æœ€é«˜é¢‘çš„ subtree sum.
* ä»€ä¹ˆå« subtree sum? The subtree sum of a node is defined as the sum of all the node values formed by the subtree rooted at that node (including the node itself). 
* è¯´ç™½äº†, å°±æ˜¯ left sub-tree + right sub-tree + node.val

##### æ€»ç»“: åŸæœ¬ä»¥ä¸ºå’Œ 501. Find Mode in Binary Search Treeæ¯”è¾ƒæ¥è¿‘, å¯ä»¥ä¸é€‚ç”¨é¢å¤–ç©ºé—´å°±å¯ä»¥æå®š.
* å®é™…ä¸Š, ä¸å¯ä»¥. 501 æœ‰ä¸€ä¸ªä¼˜åŠ¿æ¡ä»¶, å®ƒæ˜¯ BST, inorder éå†æ˜¯é€’å¢çš„. 

```python
# éœ€é¢å¤–ç©ºé—´, éœ€è¦ dic ç»Ÿè®¡é¢‘ç‡.
class Solution(object): 
    def __init__(self):
        self.maxcnt = 0
        self.dic = {}
        
    def findFrequentTreeSum(self, root):
        """
        :type root: TreeNode
        :rtype: List[int]
        """
        #1 edge condtion
        if not root: return [] #â é”™åœ¨å†™æˆ return None äº†. 
        #â æ€»ç»“: è¿”å›ç±»å‹æ˜¯ List çš„, å¦‚æœæ²¡æœ‰ä»€ä¹ˆè¿”å›, åˆ™è¿”å›ç©º List å³ []
        
        #2 get the frequencey of sums
        self.dfs(root)
        
        #3 get the results according to the dic and self.maxcnt
        rst = []
        for key, cnt in self.dic.items(): # dic.items() å®é™…ä¸Šæ˜¯ list, æ¯ä¸ªå…ƒç´ æ˜¯ tuple.
            if cnt == self.maxcnt: rst.append(key)
        return rst

    #âââ postorder éå†, éœ€è¦å…ˆå¤„ç† left å’Œ right, ç„¶ååˆå¹¶ left å’Œ right ç»“æœçš„, 
    # éƒ½éœ€è¦ post order. è¿™ä¸ªæ˜¯ bottom up ç»“æ„.
    def dfs(self, node): 
        if not node: return 0
        
        left = self.dfs(node.left)
        right = self.dfs(node.right)
        
        curSum = left + right + node.val
        self.dic[curSum] = self.dic.get(curSum, 0) + 1
        self.maxcnt = max(self.maxcnt, self.dic[curSum])
        return curSum
```

### [â¤ï¸623.medium-Add One Row to Tree](https://leetcode.com/problems/add-one-row-to-tree/description/)
##### é¢˜æ„: ç»™å®š value å’Œ depth, åœ¨ binary tree ä¸­æ’å…¥ä¸€å±‚.

```python
# æ€è·¯: 
#     1. æ‰¾åˆ° d - 1 å±‚, æ¯”å¦‚ node. å¹¶æ–°å»º left = TreeNode(v), right = TreeNode(v)
#     2. æ£€æŸ¥ node.left å’Œ node.right æ˜¯å¦å­˜åœ¨. è‹¥å­˜åœ¨, æ‹¼æ¥åˆ° left.left å’Œ right.right ä¸Šå».
#     3. å°† left å’Œ right å¯¹æ¥åˆ° node.left å’Œ node.right ä¸Šé¢.
from collections import deque
class Solution(object):
    def addOneRow(self, root, v, d):
        """
        :type root: TreeNode
        :type v: int
        :type d: int
        :rtype: TreeNode
        """
        dummy = TreeNode(None)
        dummy.left = root
        queue = deque([dummy])
        depth = 1
        while queue:
            n = len(queue)
            for i in xrange(n):
                node = queue.popleft()
                if depth == d:
                    # ----------------
                    left = TreeNode(v)
                    right = TreeNode(v)
                    if node.left: left.left = node.left
                    if node.right: right.right = node.right
                        
                    node.left = left #âââ ç†è§£é”™è¯¯,åŸæ¥æ— è®ºå¦‚ä½• node.left å’Œ node.right
                    node.right = right # éƒ½ä¼šå¾—åˆ°æ–°çš„ left node å’Œ right node.
                    continue # â¤ï¸â¤ï¸â¤ï¸âââ è¶…èµçš„ç¼–ç¨‹, å¤„ç†å®Œè¿™ä¸€å±‚, queue ä¸ºç©º, è‡ªç„¶å°±è·³å‡ºå¾ªç¯. 
                    # ----------------
                    
                if node.left: queue.append(node.left)
                if node.right: queue.append(node.right)
            depth += 1 # ââââ é”™åœ¨ depth += 1 å’Œ ä¸Šä¸€è¡Œçš„ if å¯¹é½äº†. å‡ ä¸ª for ç„å¹»è¡¨ç¤ºå‡ ä¸ª depth.
        return dummy.left
```

```python
class Solution(object):
    def trimBST(self, root, L, R):
        """
        :type root: TreeNode
        :type L: int
        :type R: int
        :rtype: TreeNode
        """
        dummy = TreeNode(L)
        dummy.left = root
        self.dfs(dummy, L, R)
        return dummy.left
        
    def dfs(self, node, L, R):
        if not node: return 
        
        if node.val < L:
            node = node.right
            node = self.dfs(node, L, R)
            return node
        if node.val > R:
            node = node.left
            node = self.dfs(node, L, R)
            return node
            
        if node.left: 
            node.left = self.dfs(node.left, L, R)
        if node.right:
            node.right = self.dfs(node.right, L, R)
        return node
```

### [366.medium-Find Leaves of Binary Tree](https://leetcode.com/problems/find-leaves-of-binary-tree/description/)
##### é¢˜æ„: Collect and remove all leaves, repeat until the tree is empty.

##### å‚è€ƒèµ„æ–™ (å€¼å¾—æŸ¥çœ‹)
* Edward Shi: https://www.youtube.com/watch?v=qJOHOKMnHdM

```python
# æ€»ç»“: è¾“å‡ºæœ‰é¡ºåº,éœ€è¦è€ƒè™‘ä¸€ä¸ª level çš„é—®é¢˜.  å› ä¸ºæ˜¯è€ƒè™‘ leaf èŠ‚ç‚¹, å› æ­¤éœ€è¦è€ƒè™‘
# æœ¬é¢˜æ˜¯ä» leaf åˆ° root æ•° level.
class Solution(object):
    def __init__(self):
        self.rst = []
        
    def findLeaves(self, root):
        """
        :type root: TreeNode
        :rtype: List[List[int]]
        """
        # edge condition
        if not root: return []
        
        # search and remove
        rst = []
        self.dfs(root, rst)
        return rst            
            
    def dfs(self, node, rst):
        # edge condition
        if not node: return -1 # â¤ï¸ return -1 éå¸¸ç‰¹åˆ«.
        
        # postorder traversal
        left = self.dfs(node.left, rst)
        right = self.dfs(node.right, rst)
        
        level = 1 + max(left, right)
        while level + 1 > len(rst):
            rst.append([])
        rst[level].append(node.val)
        node.left = node.right = None # â¤ï¸ åˆ é™¤ node
        return level    
```

### [199.medium-Binary Tree Right Side View](https://leetcode.com/problems/binary-tree-right-side-view/description/)
##### ç»™å®š Binary tree, è®¾æƒ³ä½ åœ¨ tree çš„å³è¾¹, èƒ½çœ‹åˆ°å“ªäº› node? è¯·å°†èƒ½çœ‹åˆ°çš„ nodes, ä»ä¸Šå¾€ä¸‹æ‰“å°ä¸‹æ¥.

* å‚è€ƒè§†é¢‘: https://www.youtube.com/watch?v=obLedSdUSow

```python
# æ–¹æ³•1: BFS
from collections import deque

class Solution(object):
    def rightSideView(self, root):
        """
        :type root: TreeNode
        :rtype: List[int]
        """
        if not root: return []
        
        rst, level = [], 0
        queue = deque([root])
        while queue:
            if level + 1 > len(rst): 
                rst.append(0)
                
            n = len(queue)
            for i in xrange(n):
                node = queue.popleft()
                rst[level] = node.val
                
                if node.left: queue.append(node.left)
                if node.right: queue.append(node.right)
            level += 1
        return rst
```
