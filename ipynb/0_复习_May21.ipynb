```python
class Solution(object):
    # ------------------------------------ 
    # ❤️❤️❤️❤️❤️❤️❤️
    def trapWalter(self, height):
        if not height: return 0
        
        rst = 0
        l, r = 0, len(height)-1
        maxLeft = maxRight = 0
        while l < r:
            left, right = height[l], height[r]
            if left < right:
                if left >= maxLeft:
                    maxLeft = left
                else:
                    rst += maxLeft - left
                l += 1 # ❎❎
            else:
                if right >= maxRight:
                    maxRight = right
                else:
                    rst += maxRight - right
                r -= 1 # ❎❎
        return rst

    # ------------------------------------ 
    # T & S: O(n)
    def largestRectangleArea(self, heights):
        if not heights: return 0
        
        heights.append(0)
        stack, rst = [], 0
        for i, x in enumerate(heights):
            while stack and heights[stack[-1]] > x:
                h = heights[stack.pop()]
                w = i if not stack else i-stack[-1]-1
                rst = max(rst, h*w)
            stack.append(i)
        heights.pop()
        return rst
    
    # ------------------------------------   
    # S: O(n),   T: O(n^2)
    def maximalRectangle(self, mat):
        if not mat or (len(mat)==1 and not mat[0]): return 0
        
        nr, nc = len(mat), len(mat[0])
        hist = [0] * nc
        maxArea = 0
        for r in xrange(nr):
            for c in xrange(nc):
                hist[c] = 1+hist[c] if mat[r][c]=="1" else 0
            area = self.helper7(hist)
            maxArea = max(area, maxArea)
        return maxArea
    
    def heler7(self, hist):
        if not hist: return 0
        
        hist.append(0)
        stack = []
        area = 0
        for i, x in enumerate(hist):
            while stack and hist[stack[-1]] > x:
                h = hist[stack.pop()]
                w = i if not stack else i-stack[-1]-1
                area = max(area, h*w)
            stack.append(i)
        hist.pop()
        return area
    
    # ------------------------------------ 
    # O(n^2)
    def maximalSquare(self, mat):
        if not mat or (len(mat)==1 and not mat[0]): return 0
        
        rst = 0
        nr, nc = len(mat), len(mat[0])
        dp = [[0]*(nc+1) for i in xrange(nr+1)]
        for r in xrange(1, nr+1):
            for c in xrange(1, nc+1):
                if mat[r-1][c-1] == "1":
                    dp[r][c] = 1 + min(min(dp[r-1][c], dp[r][c-1]), dp[r-1][c-1])
                    rst = max(dp[r][c], rst)
        return rst * rst
    
    
    # ------------------------------------   
    # ❤️❤️❤️❤️❤️❤️❤️
    def calculate_plusMinusParentheses(self, s):
        if not s: return 0
        
        stack = []
        l1, o1, i, n = 0, 1, 0, len(s)
        while i < n:
            x = s[i]
            if x.isdigit():
                start = i
                while i<n and s[i].isdigit(): i+=1
                val = int(s[start:i])
                i -= 1
                l1 += o1 * val
            elif x == "(":
                stack.append((l1, o1))
                l1, o1 = 0, 1
            elif x == ")":
                tmp = l1
                l1, o1 = stack.pop()
                l1 = o1 * tmp
            elif x == "+":
                o1 = 1
            elif x == "-":
                o1 = -1
            i += 1
        return l1
    
    # ------------------------------------   
    def calculate_plusMinusMultiplyDivide(self, s):
        if not s: return 0
        
        i, n = 0, len(s)
        l1, o1, l2, o2 = 0, 1, 1, 1
        while i < n:
            x = s[i]
            if x.isdigit():
                start = i
                while i < n and s[i].isdigit(): i += 1
                val = int(s[start:i])
                i -= 1 # ❤️❤️❤️❤️
                
                if o2 == 1:
                    l2 *= val
                else:
                    l2 = 1+l2/val if o2<0 and l2%val else l2/val
            elif x == "+" or x == "-":
                l1 += o1 * l2
                o1 = 1 if x=="+" else -1
                l2 = o2 = 1
            elif x == "*" or x == "/":
                o2 = 1 if x=="*" else -1
            i += 1
        return l1 + o1*l2
    
    # ------------------------------------  
    def calculate_plusMinusMultiplyDivide_parentheses(self, s):
        if not s: return 0
        
        i, n = 0, len(s)
        l1, o1, l2, o2 = 0, 1, 1, 1
        stack = []
        while i < n:
            x = s[i]
            if x.isdigit():
                start = i
                while i<n and s[i].isdigit(): i+=1 
                num = int(s[start:i])
                i -= 1
                
                l2 = l2*num if o2==1 else l2/nums
            elif x == "(":
                stack.append((l1, o1, l2, o2))
                l1, o1, l2, o2 = 0, 1, 1, 1
            elif x == ")":
                tmp = l1 + o1*l2
                l1, o1, l2, o2 = stack.pop()
                l2 = l2*tmp if o2==1 else l2/tmp
            elif x == "*" or x == "/":
                o2 = 1 if x == "*" else -1
            elif x == "+" or x == "-":
                l1 = l1 + o1 * l1
                o1 = 1 if x == "+" else -1
                o2 = l2 = 1
            i += 1
        return l1 + o1*l2
    
    # ------------------------------------ 
    # s = "3[a2[c]]", return "accaccacc".
    # s = "2[abc]3[cd]ef", return "abcabccdcdcdef".
    def decodeString(self, s):
        if not s: return ""
        
        i, n = 0, len(s)
        cnt, rst = 1, ""
        stack = []
        while i < n:
            x = s[i]
            if x.isdigit():
                start = i
                while i < n and s[i].isdigit(): i += 1
                cnt = int(s[start:i])
                i -= 1
            elif x == "[":
                stack.append((cnt, rst))
                cnt, rst = 1, ""
            elif x == "]":
                tmp = rst
                cnt, rst = stack.pop()
                rst += tmp * cnt
            else:
                rst += x
            i += 1
        return rst
                
    # ------------------------------------  
    def isValid_parentheses(self, s):
        if not s: return True
        
        stack = []
        for x in s:
            if x == "(":
                stack.append(")")
            elif x == "[":
                stack.append("]")
            elif x == "{":
                stack.append("}")
            else:
                if not stack or x != stack.pop():
                    return False
        return True if not stack else False # ❤️❤️❎❎ 很重要. 
    
    def isValid_parentheses_v2(self, s):
        if not s: return True
        
        stack = []
        dic = {"(":")", "[":"]", "{":"}"}
        for x in s:
            if x in dic:
                stack.append(dic[x])
            else:
                if not stack or stack.pop() != x:
                    return False
        return True if not stack else False
            
    # ------------------------------------   
    # minStack -------- ❤️❤️❤️
    def __init__(self):
        self.stack = []
        
    def push(self, x):
        minVal = self.getMin()
        if minVal is None or x < minVal:
            minVal = x
        self.stack.append((x, minVal))
    
    def pop(self):
        if not self.stack: 
            return None
        else:
            return self.stack.pop()
        
    def top(self):
        if not self.stack:
            return None
        else:
            return self.stack[-1][0]
        
    def getMin(self):
        if not self.stack: 
            return None
        else:
            return self.stack[-1][1]
    
    # ------------------------------------ 
    # implement stack using queue
    from collections import deque
    def __init__(self):
        self.q = deque()
        
    def push(self, x):
        self.q.append(x)
        n = len(self.q)
        for i in xrange(n-1): # ❤️❤️
            self.q.append(self.q.popleft())
        
    def pop(self):
        if not self.q:
            return None
        else:
            return self.q.popleft()
        
    def top(self):
        if not self.q:
            return 
        else:
            return self.q[0]
        
    def empty(self):
        return self.q is not None
    
    # ------------------------------------  
    # implement queue using two stacks
    def __init__(self):
        self.a = []
        self.b = []
        
    def push(self, x):
        self.a.append(x)
    
    def pop(self): # ❎❤️❤️❤️❤️
        if self.empty(): return None
        self.move()
        return self.b.pop()
            
    def top(self): # ❎❤️❤️❤️❤️
        if self.empty(): return None
        self.move()
        return self.b[-1]
        
    def move(self): # ❎❤️❤️❤️❤️
        if not self.b:
            while self.a:
                self.b.append(self.a.pop())
            
    def empty(self):
        return not self.a and self.b
        
    # ------------------------------------   
    def baseBallGamePoints(self, ops):
        if not ops: return 0
        
        rst = []
        for x in ops:
            if x == "C":
                rst.pop()
            elif x == "D":
                rst.append(2*rst[-1])
            elif x == "+":
                rst.append(rst[-1] + rst[-2])
            else:
                rst.append(int(x))
        return sum(rst)
    
    # ------------------------------------   
    def dailyTemperatures(self, temp): 
        if not temp: return []
        
        stack = []
        n = len(temp)
        rst = [0] * n
        for i in xrange(n-1, -1, -1):
            while stack and temp[i] >= temp[stack[-1]]:
                stack.pop()
            
            if stack: rst[i] = stack[-1] - i
            stack.append(i)   
        return rst
                
        
    def dailyTemperatures(self, temp):
        if not temp: return []
        
        stack = []
        n = len(temp)
        rst = [0] * n
        for i in xrange(n-1, -1, -1):
            while stack and temp[i] >= temp[stack[-1]]:
                stack.pop()
                
            if stack: rst[i] = stack[-1] - i
            
            stack.append(i)
        return rst

    
    # ------------------------------------ 
    # Input: nums1 = [4,1,2], nums2 = [1,3,4,2].  Output: [-1,3,-1]
    def nextGreaterElement(self, nums1, nums2):
        if not nums1: return []
        
        stack, dic = [], {}
        n1, n2 = len(nums1), len(nums2)
        for i in xrange(n2-1, -1, -1):
            x = nums2[i]
            while stack and x >= stack[-1]:
                stack.pop()
                
            if stack: dic[x] = stack[-1]
            stack.append(x)
            
        rst = []
        for x in nums1:
            rst.append(dic.get(x,-1))
        return rst
    
    # ------------------------------------   
    # Input: [1,2,1] Output: [2,-1,2]
    def nextGreaterElement_circle(self, nums):
        if not nums: return []
        
        n = len(nums)
        nnn = 2*n
        stack, rst = [], [-1]*n
        for i in xrange(nnn-1, -1, -1):
            x = nums[i%n]
            while stack and x >= stack[-1]:
                stack.pop()
                
            if stack and i < n: rst[i] = stack[-1]
                
            stack.append(x)
        return rst
        
    # ------------------------------------  
    """
    Some examples:
      ["2", "1", "+", "3", "*"] -> ((2 + 1) * 3) -> 9
      ["4", "13", "5", "/", "+"] -> (4 + (13 / 5)) -> 6
    """
    def evalReversePolishNotation(self, s):
        if not s: return 0
        
        stack = []
        for c in s:
            if c=="+" or c=="-" or c=="*" or c=="/":
                if not stack: return -1
                x = stack.pop()
                if not stack: return -1
                y = stack.pop()
                if c == "+":
                    tmp = x + y
                elif c == "-":
                    tmp = y - x
                elif c == "*":
                    tmp = x * y
                elif c == "/":
                    tmp = y / x
                    if tmp < 0 and y % x: tmp += 1
                stack.append(tmp)
            else:
                stack.append(int(c))
        return stack.pop()  
```

```python
class Solution(object):
    def dotProduct(self, vec1, vec2):
        if not vec1 or not vec2: return []
        
        i, j, n1, n1 = 0, 0, len(vec1), len(vec2)
        while i < n1 and j < n2:
            x, y = vec1[i], vec[j]
            if x[0] < y[0]:
                i += 1
            elif x[0] > y[0]:
                j += 1
            else:
                rst.append((x[0], x[1]*y[1]))
                i += 1
                j += 1
        return rst
    
    def dotProduct_binarySearch(self, vec1, vec2):
        if not vec1 or not vec2: return []
        if len(vec1) > len(vec2):
            self.dotProduct_binarySearch(vec2, vec1)
            
        l, n1, n2 = 0, len(vec1), len(vec2)
        for i, x in enumerate(vec1):
            j = self.binarySearch(vec2, l, n2-1, x[0])
            if j > n2-1: 
                break
            elif if j != x[0]:
                continue
            else:
                rst.append((x[0], x[1]*vec2[j][1]))
        return rst
    
    def binarySearch(self, vec2, l, r, j):
        while l < r:
            m = l + ((r-l)>>1)
            ind = vec2[m][0]
            if ind < j:
                l = m + 1
            else:
                r = m
        return l
```

```python
class Solution(object):
    # Time: O(m + n)
    def dotProduct(self, vec1, vec2):
        # vec1 中 任何一个 element 是一个 tuple, 
        # 其中 tuple 的第一个元素是 index, 第二个元素是 value.
        if not vec1 or not vec2: return []
        
        rst = []
        i, j, n1, n2 = 0, 0, len(vec1), len(vec2)
        while i < n1 and j < n2:
            x, y = vec1[i], vec2[j]
            if x[0] < y[0]:
                i += 1
            elif x[0] > y[0]:
                j += 1
            else:
                rst.append((x[0], x[1]*y[1]))
                i += 1
                j += 1
        return rst
    
    # ------------------------------------ 
    # binary search: (min(m,n))log(max(m,n))
    def dotProduct_binarySearch(self, vec1, vec2):
        if not vec1 or vec2: return []
        if len(vec1) > len(vec2): 
            return self.dotProduct_binarySearch(vec2, vec1)
        
        n1, n2 = len(vec1), len(vec2)
        for x in vec1:
            j = self.binarySearch(vec2, l, n2-1, x[0])
            if j == -1: break
                
            if j == x[0]:
                rst.append((x[0], x[1]*vec2[j][1]))
            l = j + 1 # ❤️❤️❤️👍👍👍👍
        return rst
            
    def binarySearch(self, vec, l, r, i):
        if l > r: return -1
        
        while l < r:
            m = l + ((r-l)>>1)
            ind = vec[m][0]
            if ind >= i:
                r = m
            else:
                l = m + 1
        return l
```

## Dynamic Programming: 5/27/2018

```python
class Solution(object):
    # ------------------------------------  
    def bestTimeToBuySell_1_time_buy(self, nums):
        if not nums: return 0
        
        rst = 0
        minPrice = nums[0]
        for x in nums:
            minPrice = min(minPrice, x)
            rst = max(rst, x - minPrice)
        return rst
    
    # ------------------------------------  
    def bestTimeToBuySell_multiple_trans(self, nums):
        if not nums: return 0
        
        n, rst = len(nums), 0
        for i in xrange(n-1):
            rst = max(nums[i+1]-nums[i], 0)
        return rst
    
    # ------------------------------------ 
    def bestTimeToBuySell_cooldown(self, nums):
        if not nums: return 0
        
        pre_hold = float("-inf")
        pre_sold = pre_rest = 0
        for x in nums:
            sold = pre_hold + x
            rest = max(pre_rest, pre_sold)
            hold = max(pre_hold, pre_rest - x)
            pre_hold, pre_sold, pre_rest = hold, sold, rest
        return max(sold, rest)

    # ------------------------------------ 
    def bestTimeToBuySell_TransationFee(self, nums, fee):
        if not nums: return 0
        
        pre_hold = float("-inf")
        pre_sold = pre_rest = 0
        for x in nums:
            sold = pre_hold + x - fee
            rest = max(pre_rest, pre_sold)
            hold = max(max(pre_sold, pre_rest)-x, pre_hold)
            
            pre_sold, pre_rest, pre_hold = sold, rest, hold
        return max(sold, rest) 
```

```python
class Solution(object):
    # ------------------------------------ ❤️❤️❤️❤️❤️❤️
    def maxContinuousSubArray_sum(self, nums):
        if not nums: return 0
        
        n = len(nums)
        rst = subsum = nums[0] # ××××××××
        for i in xrange(1, n): # ❤️❤️ 从 1 开始.
            subsum = max(subsum,0) + nums[i]  # ××××××××
            rst = max(rst, subsum)
        return rst
    
    # ------------------------------------ ❤️❤️❤️❤️❤️❤️
    def maxContinuousSub_product(self, nums):
        if not nums: return 0
        
        n = len(nums)
        rst = pre_max = pre_min = nums[0]
        for i in xrange(1, n): # ❤️❤️ 从 1 开始.
            x = nums[i]
            maxv = max(x, max(pre_max*x, pre_min*x))
            minv = min(x, min(pre_max*x, pre_min*x))
            rst = max(rst, max(maxv, minv))
            
            pre_max, pre_min = maxv, minv
        return rst
            
    # ------------------------------------
    # f(n) = f(n-1) + f(n-2)
    def maxCntClimbStairs(self, n):
        if n < 3: return n
        
        first, second = 1, 2
        for i in xrange(2, n): # ❤️❤️❤️
            third = first + second
            first, second = second, third
        return third
    
    # ------------------------------------ 
    # dp[n]: the min cost to climb to the n-th step # ❤️❤️❤️
    # dp[n] = min(dp[n-1]+cost[n-1], dp[n-2]+cost[n-2])
    def minCostClimbingStairs(self, cost):
        if not cost: return 0
        
        n = len(cost)
        dp = [0] * (n + 1)
        for i in xrange(2, n+1):
            dp[i] = min(dp[i-1]+cost[i-1], dp[i-2]+cost[i-2]) # 从第0步, 第1步出发...
        return dp[n]
            
    # ------------------------------------ 
    def countBits(self, n): # for i \in [0,1]
        """
        :type n: int
        :rtype: List[int]
        """
        if n < 0: return 0
        rst = [0] * (n + 1)
        for i in xrange(1, n+1):
            rst[i] = rst[i&(i-1)] + 1
        return rst
```

```python
class Solution(object):
    # ------------------------------------  
    def houseRobber_list(self, nums):
        if not nums: return 0
        
        nrob = rob = 0
        for x in nums:
            new = max(nrob, rob)
            rob = nrob + x
            nrob = new
        return max(rob, nrob)
    
    # ------------------------------------
    def houseRobber_circleList(self, nums):
        if not nums: return 0
        if len(nums) == 1: return nums[0] # ❎❎ 只有一栋房子, 则返回它里面的值.
        
        n = len(nums)
        no_1_rob = [0, 0] # ❤️ 不从第一栋开始. 0 表示不偷当前, 1 表示偷当前
        rob = [0, nums[0]] # ❤️ 从第一栋开始. 0 表示不偷当前, 1 表示偷当前
        for i in xrange(1, n):
            new = max(no_1_rob)
            no_1_rob[1] = no_1_rob[0] + nums[i]
            no_1_rob[0] = new
            
            new = max(rob)
            rob[1] = rob[0] + nums[i]
            rob[0] = new
        return max(max(no_1_rob), rob[0]) 
        
    # ------------------------------------  
    def houseRobber_binaryTree(self, root):
        if not root: return 0
        rst = self.helper(root)
        return max(rst)
        
    def helper(self, root):
        if not root: return (0, 0)
        
        left = self.helper(root.left)
        right = self.helper(root.right)
        
        nrob = max(left) + max(right)
        rob = root.val + left[0] + right[0]
        return (nrob, rob)
```

```python
import numpy as np

def initilizeCenters(X, k):
    nSmp = X.shape[0]
    return centers = X[np.random.randint(nSmp, size=k)]

def get_dist(X, center):
    return np.linalg.norm(X-center, axis=1)

def KMeans(X, k, maxIter=1000):
    nSmp, nFea = X.shape
    
    centers = initilizeCenters(X, k)
    classes = np.zeros(nSmp, dtype=np.uint16)
    dists = np.zeros((nSmp, k), dtype=np.float64)
    
    for iter in xrange(maxIter):
        for i, center in enumerate(X, centers):
            dists[:,i] = get_dist(X, center)
        classes[:] = np.argmin(dists, axis=1)
        
        for i in xrange(k):
            centers[i] = np.mean(X[classes=i,:], axis=0)
    return centers, classes
```
