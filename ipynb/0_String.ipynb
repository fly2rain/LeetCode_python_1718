### [20.easy-Valid Parentheses](https://leetcode.com/problems/valid-parentheses/description/)
##### é¢˜æ„: Given a string containing just the characters '(', ')', '{', '}', '[' and ']', determine if the input string is valid.

##### æ€»ç»“: åˆ©ç”¨ stack

```python
class Solution(object):
    def isValid(self, s):
        """
        :type s: str
        :rtype: bool
        """  
        stack = []
        for x in s:
            if x == "(":
                stack.append(")")
            elif x == "[":
                stack.append("]")
            elif x == "{":
                stack.append("}")
            else:
                if not stack or stack.pop() != x:
                    return False
        return (not stack)
```

```python
class Solution(object):
    def isValid(self, s):
        """
        :type s: str
        :rtype: bool
        """
        dic = {"(":")", "[":"]", "{":"}"}
        stack = []
        for x in s:
            if x in dic:
                stack.append(dic[x])
            elif not stack or stack.pop() != x:
                return False
        return (not stack)
```

### [22.med-Generate Parentheses](https://leetcode.com/problems/generate-parentheses/description/)
##### é¢˜æ„: Given $n$ pairs of parentheses, write a function to generate all combinations of well-formed parentheses.
##### åˆ†æ
##### æ€»ç»“

```python
class Solution(object):
    def generateParenthesis(self, n):
        """
        :type n: int
        :rtype: List[str]
        """
        if n < 1: return []
        
        rst = []
        self.helper(rst, "", n, 0, 0)
        return rst
        
    def helper(self, rst, cur, n, left, right):
        if right == n: # â¤ï¸â¤ï¸ string æ˜¯ immutable çš„, å› æ­¤ä¸ä¼ é€’çš„æ—¶å€™æ˜¯ç›´æ¥ copy çš„,
            rst.append(cur) # è€Œ list æ˜¯ mutable çš„, ä¼ å‚çš„æ—¶å€™, ç±»ä¼¼å¼•ç”¨ä¼ å‚.
        
        if left < n:
            self.helper(rst, cur + "(", n, left+1, right)
        
        if right < left:
            self.helper(rst, cur + ")", n, left, right+1)
```

### [76.hard-Minimum Window Substring](https://leetcode.com/problems/minimum-window-substring/description/)
##### é¢˜æ„: 
* Given a string S and a string T, find the minimum window in S which will contain all the characters in T in complexity $O(n)$.
* <span class="burk">Exp</span>: S = "ADOBECODEBANC", T = "ABC", Minimum window is "BANC".
* Note that: (1) å¦‚æœ S ä¸åŒ…å« T, è¾“å‡ºç©º string "".  (2) å¦‚æœå­˜åœ¨,ä¿è¯å”¯ä¸€å­˜åœ¨æ€§. 

##### è§£é¢˜ç­”æ¡ˆ https://leetcode.com/problems/minimum-window-substring/discuss/26808
* æä¾›äº†æ¸…æ™°çš„è§£é¢˜ç­”æ¡ˆ ä»¥åŠ è§£å†³è¿™ç±»é—®é¢˜çš„æ¨¡æ¿.
* é¦–å…ˆçœ‹é’ˆå¯¹è¿™é¢˜çš„ç­”æ¡ˆ, ç„¶åæŸ¥çœ‹è§£å†³è¿™ç±»é—®é¢˜çš„æ¨¡æ¿.
    * Use <span class="girk">two pointers</span>: start and end to represent a window.
    * Move end to <span class="girk">find a valid window</span>.
    * When a <span class="girk">valid window</span> is found, move start to find a smaller window.

```python
class Solution(object):
    def minWindow(self, s, t):
        """
        :type s: str
        :type t: str
        :rtype: str
        """
        # 1 edge condition
        ns, nt = len(s), len(t)
        if ns == 0 or nt == 0 or nt > ns: return ""
        
        # 2 construct the dicitonary for t
        dic = {}
        for x in t:
            dic[x] = dic.get(x,0) + 1
            
        # 3 do the sliding window search.
        cnt, j, rst = nt, 0, ""
        for i, x in enumerate(s):
            dic[x] = dic.get(x,0) - 1
            if dic[x] >= 0: cnt -= 1 # >= 0 è¡¨æ˜å·²ç»å‡ºç°è¿‡
            
            while cnt == 0:
                if i-j+1 < len(rst) or not rst:
                    rst = s[j:i+1] # â æ³¨æ„ i å’Œ j éƒ½æ˜¯ index, æ‰€ä»¥å³è¾¹ i + 1
                    
                dic[s[j]] += 1
                if dic[s[j]] > 0: cnt += 1
                j += 1
        return rst
```

```python
class Solution(object):
    def lengthOfLongestSubstringTwoDistinct(self, s):
        """
        :type s: str
        :rtype: int
        """
        if not s: return 0
        
        j, cnt, max_dist = 0, 0, 2
        rst, dic = 0, {}
        for i, x in enumerate(s):
            dic[x] = dic.get(x,0) + 1
            if dic[x] == 1: cnt += 1
                
            while cnt > max_dist:
                dic[s[j]] -= 1
                if dic[s[j]] == 0: cnt -=1
                j += 1
            rst = max(rst, i-j+1) # â¤ï¸ åŒºåˆ«åœ¨è¿™é‡Œ
                    
        return rst
```

```python
class Solution(object):
    def lengthOfLongestSubstringTwoDistinct(self, s):
        """
        :type s: str
        :rtype: int
        """
        if not s: return 0
        
        j, cnt = 0, 0
        rst, dic = 0, {}
        for i, x in enumerate(s):
            dic[x] = dic.get(x,0) + 1
            if dic[x] == 1: cnt += 1
            
            if cnt <= max_dist: rst = max(rst, i-j+1)
                
            while cnt > 2:
                dic[s[j]] -= 1
                if dic[s[j]] == 0: cnt -=1
                j += 1
                    
        return rst
```

### [438.easy-Find All Anagrams in a String](https://leetcode.com/problems/find-all-anagrams-in-a-string/description/)
##### é¢˜æ„: 
* ä¸¤ä¸ª string, æ‰¾åˆ° string 1 ä¸­æ‰€æœ‰ substring çš„ anagrams, output è¿™äº›ä½ç½®.
* ä¸¾ä¾‹ Input: s: "cbaebabacd" p: "abc", Output: [0, 6]

##### åˆ†æ: 
* æœ€æœ´ç´ çš„æƒ³æ³• compare every substrings of string s with string p.
* ä¸¤ç§æ€è·¯:
    * åˆ©ç”¨ hash-table, runtime $O(nm)$, space complexity $O(m)$, where $n$ is the size of s, $m$ is the size of p. ç«Ÿç„¶è¶…è¿‡æ—¶é—´äº†.
    * two point æ³•. runtime $O(n)$, space complexity $O(1)$, å› ä¸ºå­—æ¯æ˜¯æœ‰é™ä¸ª, ç‚¹è¿‡ 256 ä¸ª.
    
##### ç­”æ¡ˆ:
* https://www.youtube.com/watch?v=lFG63nc9zrQ
* https://leetcode.com/problems/find-all-anagrams-in-a-string/discuss/92007 é¡ºä¾¿è§£å†³ç±»ä¼¼çš„ sliding window çš„ç±»ä¼¼é—®é¢˜.
* https://www.geeksforgeeks.org/anagram-substring-search-search-permutations/
* ç–‘é—®: 
* ä½¿ç”¨ char æ¯”å¦‚'a' å¦‚ä½• index list??

```python
""" 
Example 1: ---------------------------------------------------------
Input: s: "cbaebabacd" p: "abc"
Output: [0, 6]

Explanation:
The substring with start index = 0 is "cba", which is an anagram of "abc".
The substring with start index = 6 is "bac", which is an anagram of "abc".

Example 2: ---------------------------------------------------------
Input: s: "abab" p: "ab"
Output: [0, 1, 2]

Explanation:
The substring with start index = 0 is "ab", which is an anagram of "ab".
The substring with start index = 1 is "ba", which is an anagram of "ab".
The substring with start index = 2 is "ab", which is an anagram of "ab".
"""
```

```python
class Solution(object):
    def findAnagrams(self, s, p):
        """
        :type s: str
        :type p: str
        :rtype: List[int]
        """
        # 1 edge condition
        ns, np = len(s), len(p)
        if ns == 0 or np == 0 or ns < np:
            return []
        
        # 2 construct the template dicitonary
        dic = {}
        for x in p: 
            dic[x] = dic.get(x,0) + 1 
         
        # 3 do the compare
        cnt, j = np, 0
        rst = []
        for i, x in enumerate(s):
            if i-j+1 > np: # â¤ï¸â¤ï¸ keep the sliding window size is exactly as the length of p
                dic[s[j]] += 1
                if dic[s[j]] > 0: cnt += 1
                j += 1
                    
            dic[x] = dic.get(x,0) - 1
            if dic[x] >= 0:
                cnt -= 1
            
            if cnt == 0:
                rst.append(j)
        return rst     
```

### [14.easy-Longest Common Prefix](https://leetcode.com/problems/longest-common-prefix/description/)
##### é¢˜æ„: write a function to <span class="girk">find the longest common prefix string</span> amongst an array of strings.

```python
# æ–¹æ³•çœ‹ç€æ›´å¥½ä¸€ç‚¹, æ²¡æœ‰äº† min(strs, key=len)
class Solution(object):
    def longestCommonPrefix(self, strs):
        """
        :type strs: List[str]
        :rtype: str
        """
        if not strs: return ""
        
        prefix = strs[0]
        for i, x in enumerate(prefix): # â ç«Ÿç„¶å°† prefix å†™æˆäº† strs
            for s in strs:
                if i < len(s) and s[i] == x:
                    continue
                else:
                    return prefix[:i]
        return prefix

print Solution().longestCommonPrefix(["aa", "aa"])
```
